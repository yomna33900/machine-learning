```python
df['Trade Value (US$)'] = df['Trade Value (US$)'].astype(object)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df.columns = df.columns.str.strip()

Partner_values = df.groupby('Partner')['Trade Value (US$)'].sum().reset_index()

Partner_values = Partner_values.sort_values(by='Trade Value (US$)', ascending=False).head(10)

sns.barplot(data=Partner_values, x='Trade Value (US$)', y='Partner')
plt.title("Top 10 Exporting Countries by Total Trade Value (US$)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df.columns = df.columns.str.strip()

Commodity_values = df.groupby('Commodity')['Trade Value (US$)'].sum().reset_index()

Commodity_values = Commodity_values.sort_values(by='Trade Value (US$)', ascending=False).head(10)

sns.barplot(data=Commodity_values, x='Trade Value (US$)', y='Commodity')
plt.title("Top 10 Exporting Countries by Total Trade Value (US$)")
plt.tight_layout()
plt.show()
```
